```python
bar = sc.Polygon(points=sc.geometry.box(10, 2))
plus = bar.union(bar.rotate(90))

mesh = plus.make_mesh(min_points=1500)
x, y = mesh.sites.T

z_hat = np.array([0, 0, 1])  # unit vector in the z direction
magnetization = sc.ureg("1 mu_B / nm ** 2")

vertex_positions = np.stack([x, y, np.zeros_like(x)], axis=1)
vertex_areas = mesh.vertex_areas * sc.ureg("um ** 2")
vertex_moments = z_hat * magnetization * vertex_areas[:, np.newaxis]
# Convert to units of Bohr magnetons
vertex_moments = vertex_moments.to("mu_B").magnitude

field = sc.sources.DipoleField(
    dipole_positions=vertex_positions,
    dipole_moments=vertex_moments,
)

# Coordinates at which to evaluate the magnetic field (in microns)
N = 101
eval_xs = eval_ys = np.linspace(-6, 6, N)
eval_z = 0.5
xgrid, ygrid, zgrid = np.meshgrid(eval_xs, eval_ys, eval_z)
xgrid = np.squeeze(xgrid)
ygrid = np.squeeze(ygrid)
zgrid = np.squeeze(zgrid)

# field returns shape (N * N, 3), where the last axis is field component
# and the units are Tesla
Hz = field(xgrid.ravel(), ygrid.ravel(), zgrid.ravel()) * sc.ureg("tesla")
# Reshape to (N, N, 3), where the last axis is field component,
# and convert to microTesla
Hz = Hz.reshape((N, N, 3)).to("microtesla").magnitude
```

```python
fig, axes = plt.subplots(
    1, 3, figsize=(6, 2), sharex=True, sharey=True, constrained_layout=True
)

vmax = max(abs(Hz.min()), abs(Hz.max()))
vmin = -vmax
kwargs = dict(vmin=vmin, vmax=vmax, shading="auto", cmap="cividis")

for i, (ax, label) in enumerate(zip(axes, "xyz")):
    ax.set_aspect("equal")
    im = ax.pcolormesh(xgrid, ygrid, Hz[..., i], **kwargs)
    ax.set_title(f"$\\mu_0H_{{{label}}}(z={{{eval_z}}}\\,\\mu\\mathrm{{m}})$")
cbar = fig.colorbar(im, ax=axes)
cbar.set_label("$\\mu_0H$ [$\\mu$T]")
_ = axes[0].set_xlabel("$x$ [$\\mu$m]")
_ = axes[0].set_ylabel("$y$ [$\\mu$m]")
```

```python
import superscreen as sc
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the ring-with-slit geometry
ro = 300  # outer radius
ri = 50  # inner radius
slit_width = 5
# layer = sc.Layer("base", london_lambda=0.100, thickness=0.025, z0=0)
geometry = sc.geometry.box(
    sc.geometry.box(ro),
    sc.geometry.box(ri),
    sc.geometry.box(slit_width, 1.5 * (ro - ri), center=(0, -(ro + ri) / 2))
)
film = sc.Film("ring_with_slit", geometry=geometry)
device = sc.Device("ring_device", films=[film])
device.make_mesh(min_points=750)

# Step 2: Define the dipole field source
dipole = sc.sources.DipoleField(
    dipole_positions=(0, 0, 0),
    dipole_moments=(0, 0, 1),
    moment_units="uA * um ** 2",
    length_units="um",
)

# Add the field to the device
device.applied_field = dipole

# Step 3: Solve the device response
solution = device.solve()

# Step 4: Evaluate the magnetic field at a grid of points
N = 101
eval_xs = eval_ys = np.linspace(-150, 150, N)
eval_z = 5  # microns above the plane of the SQUID
xgrid, ygrid, zgrid = np.meshgrid(eval_xs, eval_ys, eval_z)
xgrid = np.squeeze(xgrid)
ygrid = np.squeeze(ygrid)
zgrid = np.squeeze(zgrid)

# Step 5: Evaluate the total field (dipole + screening)
# You can also evaluate only the screening using solution.film_fields[film]
total_field = solution.total_field(
    xgrid.ravel(), ygrid.ravel(), zgrid.ravel()
) * sc.ureg("tesla")

Hz = total_field.reshape((N, N, 3))[:, :, 2].to("microtesla").magnitude

# Step 6: Plot the z-component of the field
plt.figure(figsize=(6, 5))
plt.contourf(eval_xs, eval_ys, Hz, levels=100, cmap="RdBu_r")
plt.colorbar(label="Hz [µT]")
plt.xlabel("x [µm]")
plt.ylabel("y [µm]")
plt.title("Total B-field (Hz component) from dipole + screening")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
import superscreen as sc
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Setup units & Layer
length_units = "um"
layer = sc.Layer("base", london_lambda=0.100, thickness=0.025, z0=0)

# Step 2: Build ring-with-slit geometry manually
ro = 150; ri = 40; slit_w = 4
ring = sc.geometry.circle(ro)
hole = sc.geometry.circle(ri)
slit = sc.geometry.box(slit_w, 1.5 * (ro - ri), center=(0, -(ro + ri) / 2))
film = sc.Polygon.from_difference([ring, hole, slit], name="ring", layer="base")

# Step 3: Create Device and build mesh
device = sc.Device("ring_dev", layers=[layer], films=[film], length_units=length_units)
device.make_mesh(min_points=1500)

# Step 4: Define the dipole source
dipole = sc.sources.DipoleField(
    dipole_positions=(0, 0, 0),
    dipole_moments=(0, 0, 1),
    moment_units="uA * um ** 2",
    length_units=length_units,
)

# Step 5: Set up the solver
solver = sc.Solver(device=device, applied_field=dipole)

# Step 6: Run the solver
solution = solver.solve()  # <- This computes screening currents & fields

# Step 7: Compute total B-field (source + screening) at z = 5 µm
N = 101
xs = ys = np.linspace(-200, 200, N)
xg, yg, zg = np.meshgrid(xs, ys, 5)
x, y, z = map(np.ravel, (xg, yg, zg))
B = solution.total_field(x, y, z) * sc.ureg("tesla")
Hz = B[:, 2].reshape((N, N)).to("microtesla").magnitude

# Step 8: Visualize the Hz component
plt.figure(figsize=(6, 5))
plt.contourf(xs, ys, Hz, levels=100, cmap="RdBu_r")
plt.colorbar(label="Hz [µT]")
plt.xlabel("x [µm]")
plt.ylabel("y [µm]")
plt.title("Total Hz from dipole + screening")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
%config InlineBackend.figure_formats = {"retina", "png"}
%matplotlib inline

import os
import numpy as np
import matplotlib.pyplot as plt
import superscreen as sc
from superscreen.geometry import box

# Set up matplotlib
os.environ["OPENBLAS_NUM_THREADS"] = "1"
plt.rcParams["figure.figsize"] = (4, 4)
plt.rcParams["font.size"] = 10

# Check Superscreen version
sc.version_table()

# === Film + Slot + Center Hole ===
width = 600
height = 620
slot_width = 250
slot_height = 20
hole_size = 100  # Square hole: 100 um x 100 um

center = (0, 0)
length_units = "um"

layer = sc.Layer("base", Lambda=0.9)

# Create film
film = sc.Polygon("film", layer="base", points=box(width, height))

# Side slots
slot = sc.Polygon(points=box(slot_width, slot_height, center=(-175, -305)))
film = film.difference(slot, slot.scale(xfact=-1))

# Center square hole
hole = sc.Polygon(points=box(hole_size, hole_size, center=center))
film = film.difference(hole)

# Optional: Visualize film
# film.plot()
# plt.title("Film with Center Hole")
# plt.show()

# === Dipole at Center ===
dipole_position = np.array([[0.0, 0.0, 0.0]])  # Inside hole
dipole_moment = np.array([[0, 0, 1]]) * sc.ureg("1 mu_B")  # z-direction
dipole_moment = dipole_moment.to("mu_B").magnitude

field = sc.sources.DipoleField(
    dipole_positions=dipole_position,
    dipole_moments=dipole_moment,
)

# === Field Evaluation Grid ===
N = 101
eval_xs = eval_ys = np.linspace(-6, 6, N)
eval_z = 0.5
xgrid, ygrid, zgrid = np.meshgrid(eval_xs, eval_ys, eval_z)
xgrid = np.squeeze(xgrid)
ygrid = np.squeeze(ygrid)
zgrid = np.squeeze(zgrid)

Hz = field(xgrid.ravel(), ygrid.ravel(), zgrid.ravel()) * sc.ureg("tesla")
Hz = Hz.reshape((N, N, 3)).to("microtesla").magnitude

# === Plot Field and Dipole ===
fig, axes = plt.subplots(
    1, 3, figsize=(6, 2), sharex=True, sharey=True, constrained_layout=True
)

vmax = max(abs(Hz.min()), abs(Hz.max()))
vmin = -vmax
kwargs = dict(vmin=vmin, vmax=vmax, shading="auto", cmap="cividis")

for i, (ax, label) in enumerate(zip(axes, "xyz")):
    ax.set_aspect("equal")
    im = ax.pcolormesh(xgrid, ygrid, Hz[..., i], **kwargs)
    ax.set_title(f"$\\mu_0H_{{{label}}}(z={{{eval_z}}}\\,\\mu\\mathrm{{m}})$")

    # Plot dipole position
    ax.plot(
        dipole_position[:, 0],
        dipole_position[:, 1],
        "ro",
        markersize=8,
        markeredgecolor="k",
        markeredgewidth=1,
    )

cbar = fig.colorbar(im, ax=axes)
cbar.set_label("$\\mu_0H$ [$\\mu$T]")
axes[0].set_xlabel("$x$ [$\\mu$m]")
axes[0].set_ylabel("$y$ [$\\mu$m]")
plt.show()
```
